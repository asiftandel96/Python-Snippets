#                                            FEATURE ENGINEERING

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('train.csv')
```

```python
dataset.head()
```

```python
## Let us capture all the nan values
## First lets handle Categorical features which are missing
feature_withnan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes=='O']

for feature in feature_withnan:
    print("{}: {}% missing values".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
# Capture nan value with the missing label
def replace_cat_feature(dataset,feature_withnan):
    data=dataset.copy()
    data[feature_withnan]=data[feature_withnan].fillna('Missing')
    return data

dataset=replace_cat_feature(dataset,feature_withnan)
dataset[feature_withnan].isnull().sum()
```

```python
dataset.head()
```

```python
## Now lets check for numerical variables the contains missing values
numerical_withnan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtypes!='O']
## We will print the numerical nan variables and percentage of missing values
for feature in numerical_withnan:
    print("{} {}%missing values".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
## Replacing the numerical Missing Values
for feature in numerical_withnan:
    ## We will replace by using median since there are outliers
    median_value=dataset[feature].median()
    ## create a new feature to capture nan values
    dataset[feature+'nan']=np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value,inplace=True)
    
dataset[numerical_withnan].isnull().sum()
    
```

```python
dataset.head()
```

```python

## Temporal Variables (Date Time Variables)
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    
    dataset[feature]=dataset['YrSold']-dataset[feature]
```

```python
dataset.head()
```

```python
dataset[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

NUMERICAL VARIABLES


Since the numerical variables are skewed we will perform log normal distribution

```python
dataset.head()
```

```python
num_feature=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for feature in num_feature:
    dataset[feature]=np.log(dataset[feature])
```

```python
dataset[num_feature].head()
```

Handling Rare Categorical Feature¶

We will remove categorical variables that are present less than 1% of the observations

```python
categorical_feature=[feature for feature in dataset.columns if dataset[feature].dtype=='O']
print(categorical_feature)
```

```python
for feature in categorical_feature:
    temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df=temp[temp>0.01].index
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
```

```python
dataset.head(10)
```

```python
for feature in categorical_feature:
    label_ordered=dataset.groupby(feature)['SalePrice'].mean().sort_values().index
    label_ordered={k:i for i,k in enumerate(label_ordered,0)}
    dataset[feature]=dataset[feature].map(label_ordered)
```

```python
dataset.head()
```

```python
scaling_feature=[feature for feature in dataset.columns if feature not in ['Id','SalePrice']]
len(scaling_feature)
```

```python
print(scaling_feature)
```

FEATURE SCALING

```python
from sklearn.preprocessing import MinMaxScaler
Scaler=MinMaxScaler()
Scaler.fit(dataset[scaling_feature])
```

```python
Scaler.transform(dataset[scaling_feature])
```

```python

# transform the train and test set, and add on the Id and SalePrice variables
data = pd.concat([dataset[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(Scaler.transform(dataset[scaling_feature]), columns=scaling_feature)],
                    axis=1)
```

```python
data.head()
```

```python
data.to_csv('X_train.csv',index=False)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor    

def calculate_vif_(dataset_price, thresh=5.0):
    variables = list(range(dataset_price.shape[1]))
    dropped = True
    while dropped:
        dropped = False
        vif = [variance_inflation_factor(dataset_price.iloc[:, variables].values, ix)
               for ix in range(dataset_price.iloc[:, variables].shape[1])]

        maxloc = vif.index(max(vif))
        if max(vif) > thresh:
            print('dropping \'' + dataset_price.iloc[:, variables].columns[maxloc] +
                  '\' at index: ' + str(maxloc))
            del variables[maxloc]
            dropped = True

    print('Remaining variables:')
    print(dataset_price.columns[variables])
    return dataset_price.iloc[:, variables]
```
