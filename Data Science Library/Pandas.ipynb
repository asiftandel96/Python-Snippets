# ####WORKING WITH DATETIME AND TIME SERIES DATA

```python
#Now we are working with time series data.Here we are converting it into datetime format by using format as seen 
# from the python datetime documentation from the python.org website
# Here '%Y-Year %m-month %d-Day %I-24 hour clock %p-AM OR PM
d_parser=lambda x:pd.datetime.strptime(x,'%Y-%m-%d %I-%p')
data_time=pd.read_csv('ETH_1h.csv',parse_dates=['Date'],date_parser=d_parser)
```

```python
# Checking if we have applied our function correctly by seeing the first 18 results 
data_time.head(18)
```

```python
#Show the date day
data_time.loc[0,'Date'].day_name()
```

```python
# Showing the day_name of the whole dataframe(Note : using datetime class)
data_time['Date'].dt.day_name()
```

```python
# Making different columns for the day_name
data_time['DayOfWeek']=data_time['Date'].dt.day_name()
data_time
```

```python
# Last Date
data_time['Date'].min()
```

```python
# First Date
data_time['Date'].max()
```

```python
# Last Day-First Day-TimeDelta
data_time['Date'].max() - data_time['Date'].min()
```

```python
# Filtering out the date between >Jan 2019 and < Jan 2020 
Filt=(data_time['Date']>=pd.to_datetime('2019-01-01')) & (data_time['Date']<pd.to_datetime('2020-01-01'))
data_time.loc[Filt]
```

```python
# Now setting Date as the index and permanently stored in the dataframe by using (inplace=True)
data_time.set_index('Date',inplace=True)
```

```python
data_time.head(50)
```

```python
# Checking if the Date is applied as the Index
data_time.loc['2019-01-01']
```

```python

data_time.loc['2020-01':'2020-02']['Close'].mean()
```

```python

data_time.loc['2020-01-01']['High'].max()
```

```python
# If we want to change the hours retreival to day retreival of the data then we can use resample('D') 
Highs=data_time.resample('D').max()
```

```python
# Applying the concepts of resample and seeing the result
Highs.loc['2020-01-01']
```

```python
%matplotlib inline
```

```python

Highs.plot()
```

```python
# Here we have used resample('W')
Every_Week=data_time.resample('W')['High'].mean()
```

```python
# We can also use agg function to change the metrics of every column(i.e metric meaning some columns need min and some need median)
# and elsewhere
data_time.resample('W').agg({'Close':'mean','High':'max','Low':'min','Volume':'sum'})
```

```python
%matplotlib inline
```

```python
Every_Week.plot()
```
